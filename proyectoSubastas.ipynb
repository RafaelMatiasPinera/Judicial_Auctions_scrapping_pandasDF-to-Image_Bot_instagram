```python
######################el bueno###################################
from selenium import webdriver
from selenium.webdriver.firefox.options import Options
def html_to_image(html_file, filename, width, height):
    """Converts an HTML file to an image using Selenium and Python with a specified width and height."""

    options = Options()
    options.add_argument('--headless')
    browser = webdriver.Firefox(options=options)
    
    # Set the window size to the desired width and height
    browser.set_window_size(width, height)
    
    with open(html_file, 'r', encoding='utf-8') as f:
        html_element = f.read()

    # Modify the HTML content to make the table larger and add a title
    modified_html = modify_html_for_instagram(html_element)
    
    browser.get('data:text/html;charset=utf-8,' + modified_html)

    browser.save_screenshot(filename)

def modify_html_for_instagram(original_html):
    # Modify the HTML to make the table larger and add a title
    modified_html = original_html.replace('<table', '<table width="100%" style="font-size: 24px; padding: 20px;"')
    modified_html = modified_html.replace('<td', '<td style="width: 15%; text-align: center;"')
    
    title = '<h1 style="text-align: center;">Subastas del Día de Hoy</h1>'
    
    return f"{title}{modified_html}"

html_file = 'output.html'
filename = 'data.png'

# Adjust the width and height as needed to achieve 1080x1080 px size
width = 1080
height = 1080

html_to_image(html_file, filename, width, height)
```
